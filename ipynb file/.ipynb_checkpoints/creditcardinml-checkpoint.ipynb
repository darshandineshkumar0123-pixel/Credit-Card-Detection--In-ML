Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Create sample credit card fraud dataset
import numpy as np
import pandas as pd
np.random.seed(42)


# Generate sample data similar to credit card fraud dataset
n_samples = 5000
n_legit = int(n_samples * 0.98)
n_fraud = n_samples - n_legit

# Create features
features = np.random.randn(n_samples, 28)

# Create class labels (0 = legitimate, 1 = fraud)
labels = np.concatenate([np.zeros(n_legit), np.ones(n_fraud)])

# Create DataFrame
credit_card_data = pd.DataFrame(features, columns=[f'V{i}' for i in range(1, 29)])
credit_card_data['Amount'] = np.random.exponential(50, n_samples)
credit_card_data['Class'] = labels
credit_card_data = credit_card_data.sample(frac=1).reset_index(drop=True)

print("Dataset created successfully!")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better-looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 10)

# Create subplots for visualizations
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# 1. Class Distribution - Bar Chart
class_counts = credit_card_data['Class'].value_counts()
axes[0, 0].bar(['Legitimate (0)', 'Fraudulent (1)'], class_counts.values, color=['green', 'red'])
axes[0, 0].set_title('Class Distribution - Transaction Types', fontsize=12, fontweight='bold')
axes[0, 0].set_ylabel('Number of Transactions')
for i, v in enumerate(class_counts.values):
    axes[0, 0].text(i, v + 50, str(v), ha='center', fontweight='bold')

# 2. Class Distribution - Pie Chart
axes[0, 1].pie(class_counts.values, labels=['Legitimate', 'Fraudulent'], autopct='%1.2f%%', 
               colors=['green', 'red'], startangle=90)
axes[0, 1].set_title('Transaction Distribution Percentage', fontsize=12, fontweight='bold')

# 3. Amount Distribution by Class - Box Plot
credit_card_data.boxplot(column='Amount', by='Class', ax=axes[1, 0])
axes[1, 0].set_title('Transaction Amount Distribution by Class', fontsize=12, fontweight='bold')
axes[1, 0].set_xlabel('Class (0=Legitimate, 1=Fraudulent)')
axes[1, 0].set_ylabel('Amount ($)')
plt.sca(axes[1, 0])
plt.xticks([1, 2], ['Legitimate (0)', 'Fraudulent (1)'])

# 4. Histogram - Amount Distribution
axes[1, 1].hist(credit_card_data[credit_card_data['Class'] == 0]['Amount'], bins=50, 
                alpha=0.6, label='Legitimate', color='green')
axes[1, 1].hist(credit_card_data[credit_card_data['Class'] == 1]['Amount'], bins=50, 
                alpha=0.6, label='Fraudulent', color='red')
axes[1, 1].set_title('Transaction Amount Distribution', fontsize=12, fontweight='bold')
axes[1, 1].set_xlabel('Amount ($)')
axes[1, 1].set_ylabel('Frequency')
axes[1, 1].legend()

plt.tight_layout()
plt.show()

print("Visualization Summary:")
print(f"Total Transactions: {len(credit_card_data)}")
print(f"Legitimate Transactions: {(class_counts[0])} ({class_counts[0]/len(credit_card_data)*100:.2f}%)")
print(f"Fraudulent Transactions: {(class_counts[1])} ({class_counts[1]/len(credit_card_data)*100:.2f}%)")
```

## Model Training

Using Logistic Regression to train the model

```python
# Create correlation heatmap and additional visualizations
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# 1. Correlation Heatmap - Top features with Class
correlation = credit_card_data.corr()
class_correlation = correlation['Class'].sort_values(ascending=False)[1:11]
sns.barplot(x=class_correlation.values, y=class_correlation.index,hue=class_correlation.index, ax=axes[0],legend=False)
axes[0].set_title('Top 10 Features Correlated with Class', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Correlation Coefficient')

# 2. Legitimate vs Fraudulent Mean Comparison
comparison_means = credit_card_data.groupby('Class')[['Amount']].mean()
axes[1].bar(['Legitimate', 'Fraudulent'], comparison_means['Amount'].values, color=['green', 'red'])
axes[1].set_title('Average Transaction Amount by Class', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Average Amount ($)')
for i, v in enumerate(comparison_means['Amount'].values):
    axes[1].text(i, v + 1, f'${v:.2f}', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# training the Logistic Regression Model with Training Data
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('logreg', LogisticRegression(max_iter=1000))
])

pipe.fit(X_train, Y_train)
```

```python
# Make predictions on training and test data
X_train_prediction = pipe.predict(X_train)
X_test_prediction = pipe.predict(X_test)

# Calculate accuracy scores
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)

print('Accuracy on Training data : ', training_data_accuracy)
print('Accuracy on Test data : ', test_data_accuracy)
```

```python
# Model Performance Visualization
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Create confusion matrix
cm = confusion_matrix(Y_test, X_test_prediction)

# Plot confusion matrix
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Confusion Matrix Heatmap
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0], 
            xticklabels=['Legitimate', 'Fraudulent'],
            yticklabels=['Legitimate', 'Fraudulent'])
axes[0].set_title('Confusion Matrix - Test Data', fontsize=12, fontweight='bold')
axes[0].set_ylabel('True Label')
axes[0].set_xlabel('Predicted Label')

# Model Accuracy Comparison
accuracies = [training_data_accuracy, test_data_accuracy]
dataset_types = ['Training Data', 'Test Data']
colors = ['#2ecc71', '#3498db']
bars = axes[1].bar(dataset_types, accuracies, color=colors, width=0.6)
axes[1].set_title('Model Accuracy Comparison', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Accuracy Score')
axes[1].set_ylim([0, 1])
for bar, acc in zip(bars, accuracies):
    height = bar.get_height()
    axes[1].text(bar.get_x() + bar.get_width()/2., height,
                f'{acc:.4f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# Print Classification Report
print("\nClassification Report (Test Data):")
print(classification_report(Y_test, X_test_prediction, target_names=['Legitimate', 'Fraudulent'], zero_division=0))
```

```python
from flask import Flask, request, jsonify

app = Flask(__name__)

@app.route('/')
def home():
    return "Fraud Detection API Running"

@app.route('/predict', methods=['POST'])
def predict():
    data = request.get_json(force=True)
    if not data or 'features' not in data:
        return jsonify({
            "error": "JSON payload must include a 'features' list of 29 numeric values."
        }), 400

    features = data['features']
    if not isinstance(features, (list, tuple)) or len(features) != 29:
        return jsonify({
            "error": "'features' must be a list or tuple of 29 numeric values."
        }), 400

    prediction = pipe.predict([features])[0]

    return jsonify({
        "prediction": int(prediction),
        "result": "Fraud" if prediction == 1 else "Legitimate"
    })

# To run Flask inside notebook (optional)
# app.run(debug=True)
```
